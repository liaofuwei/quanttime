基准中证证券指数 399975.XSHE	中证全指证券公司指数(四级行业)	ZQGS	2013-07-15	2200-01-01	index

```python
#从joinquant获取中证证券指数作为基准，单个指数数据，实时获取，暂不维护数据
today = datetime.today().date().strftime("%Y-%m-%d")
security_index_k = get_price("399975.XSHE", start_date='2013-07-15', end_date=today, frequency='daily', fields=None)
security_index_k = security_index_k.dropna()
security_index_k.head(4)
```

```python
secu_index_kline = Kline()
secu_index_bar = Bar()
secu_index_datetime = np.array(security_index_k.index.map(str))
ohlc = np.array(security_index_k.loc[:,['open','close','low','high']])
vol = np.array(security_index_k.loc[:,["volume"]])

secu_index_kline.add('399975',secu_index_datetime,ohlc,mark_point=["max","min"],is_datazoom_show=True,datazoom_orient='horizontal')
secu_index_bar.add('399975',secu_index_datetime,vol,is_datazoom_show=True,datazoom_xaxis_index=[0,1])

secu_index_grid = Grid()
secu_index_grid.add(secu_index_bar,grid_top="80%")
secu_index_grid.add(secu_index_kline,grid_bottom="30%")
secu_index_kline
```
