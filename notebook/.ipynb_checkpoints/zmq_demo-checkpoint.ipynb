```python
import tushare as ts
import zmq
```

```python
cons = ts.get_apis()
bond_realtime_price = ts.quotes(["110030","113008"],conn=cons)
bond_realtime_price

```

```python
context   = zmq.Context()
publisher = context.socket(zmq.PUB)
publisher.bind("tcp://*:5563")
```

```python
bond_realtime_price.columns
```

```python
sendmsg = ["code","110030","price","1029.0","last_close","1028.6"]
publisher.send_multipart(sendmsg)
```

```python

subscriber = context.socket(zmq.SUB)   
subscriber.connect("tcp://localhost:5563")   
subscriber.setsockopt_string(zmq.SUBSCRIBE,"")

while True:   
    revmsg = subscriber.recv_multipart()
    print(revmsg)
```
